## Evaluation of Models 

**Our Models:**

We decided to compare the findings using models such a DBSCAN, K-Means Clustering and Hierachical Clustering within out model, and later compared our findings to find the best model for emmission reduction optimization. We picked a unsupervised learning approach since we had unlabelled data where we were attempthing to find patterns of CO2 reduction and ride pooling. Within this we chose to explore K-Means and Hierarchical Clustering since they were topics coverd during the MSE 446 course and we wished to utilize them within our project to enhance our course understanding. Additionally they seemed like vailable options to try to find emmission reduction via pooling due to the idea of clustering similar rides. We picked DBScan since it seemed to provide a good alternatove to the 2 models learnt in class to fuether deepen our Machine Learning experience. 

**Test and Validation Approach:**

In order to validate our models, we decided to use a 10,000 row test subset to train the DBScan, K-Means and Hierarchical models. Additionally we used another 10,000 rows of data from our initial dataset of the City of New York's OpenData Portal to validate our findings. 

We were only able to use 10,000 rows of data to test and validate our findings, even though we had over 2 million rows of data, as seen in our models since we were unable to run more than 10,000 rows on our Jupyter Notebooks, which is the method of delivery taught in this course. If we attempted to use more than 10,000 rows our Python Kernel within the GitHub codespaces would crash, preventing us from moving forward with our model creation and testing. This was the reason behind the choice of 10,000 rows of data being used for training and Validation

**Findings:**

Our findings of the data can be found within model_comparison.ipynb and poolability_scaleup.ipynb. Within our poolability analysis, it was found that DBSCAN had the best perfeormance regarding CO2 % decreased with a 1% decrease as compared to K-means with a 0.6% reduction and Hierarchical clustering with a 0.5% reduction.
